```python
import time, math, csv, datetime, pandas
import matplotlib.pyplot as plt
from datetime import date 
from skimpy import skim
import lightgbm as lgb
```

```python
def genderchange(df):
    df['gender'].replace(to_replace= 'M', value = True, inplace=True )
    df['gender'].replace(to_replace= 'F', value = False, inplace=True )
```

```python
def cat_to_num(df,dfref):
    job_dictionary = {}
    for x, i in enumerate(df[dfref].values):
        if i not in job_dictionary.values(): 
            job_dictionary[x] = i
    job_df = pandas.DataFrame.from_dict(job_dictionary, orient='index', columns=[dfref])
    print(job_df)
    for x in job_df[dfref]:
        ix = job_df.index[job_df[dfref] == x][1]
        df[dfref].replace(to_replace= [x], value = ix , inplace=True )
        print(int(ix) + str(x))
```

```python
df_train = pandas.read_csv('fraudtrain.csv', header=0)
df_test = pandas.read_csv('fraudtest.csv', header=0)
```

```python
#genderchange(df_train)
#genderchange(df_test)
cat_to_num(df_train,'job')
```

```python
features = ['lat', 'long', 'amt', 'merch_long', 'merch_lat']
train_data = lgb.Dataset(df_train[features],feature_name = 'auto')
test_data = lgb.Dataset(df_test[features] , feature_name = 'auto')
train_data
```

```python
params = {'objective': 'binary', 'metric': 'auc', 'learning_rate': 0.1, 'num_leaves': 31}

model = lgb.train(params, train_data, valid_sets=test_data, label='is_fraud')
```
